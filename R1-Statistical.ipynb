```python
#Importing required libraries

import numpy as np
import pandas as pd
```

```python
# import the data from csv

df = pd.read_csv('Laliga.csv')
df
```

```python
# The header is coming as unnamed and actual header is in second row. Therefore restoring df from 1st row

df.columns = df.iloc[0]
df = df[1:]
```

```python
# replacing the values of dash with 0 

df=df.replace('-',0)
```

```python
df
```

```python
# There are year values like 1930 and 1930-31 in Debut column so I am creating another column with year values on which I will apply condition

df['DebutYear']=df.Debut.str.split('-').str.get(0).astype(int)
```

```python
# Print all the teams which have started playing between 1930-1980. 
df[(df.DebutYear>=1930) & (df.DebutYear<=1980)]
```

```python
# Points is not in numeric so converting it to numeric before I apply the sort on it
df['Points']=pd.to_numeric(df['Points'])
```

```python
df.sort_values('Points',ascending=False).head(5)
```

```python
#Goal difference function
def GoalDiff(f,a):
    return f-a
```

```python
#Passing goalsFor and GoalsAgainst columns from the df to function to get all teams goaldifference.

df['GoalDifference']=GoalDiff(pd.to_numeric(df.GoalsFor),pd.to_numeric(df.GoalsAgainst))
```

```python
df[['Team','GoalDifference']]
```

```python
#Find team with max goal difference. I am using function directly here. We can use the newly created column too

df.loc[GoalDiff(pd.to_numeric(df.GoalsFor),pd.to_numeric(df.GoalsAgainst)).idxmax()]
```

```python
#Find team with min goal difference. I am using function directly here. We can use the newly created column too

df.loc[GoalDiff(pd.to_numeric(df.GoalsFor),pd.to_numeric(df.GoalsAgainst)).idxmin()]
```

```python
# create column Winning Percent

df["Winning Percent"]=(pd.to_numeric(df.GamesWon) / pd.to_numeric(df.GamesPlayed))*100
```

```python
df["Winning Percent"].fillna(0, inplace=True)
```

```python
# Top 5 teams with winning percentage
df.sort_values("Winning Percent",ascending=False).head(5)
```

```python
# Group on BestPosition and sum of points

df.groupby('BestPosition')['Points'].agg('sum')
```
